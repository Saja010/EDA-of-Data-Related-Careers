# The Top 3 Jobs in Salary

```python
Top3_jobs=df.sort_values(by=['salary_in_usd'], ascending=False).head(3)
```

```python
Top3_jobs
```

```python
# Plotting a bar chart

plt.figure(figsize=(15, 5))
plt.bar(Top3_jobs['job_title'], Top3_jobs['salary_in_usd'],width = 0.6)


# Adding labels and title
plt.xlabel('job_title')
plt.ylabel('salary_in_usd')
plt.title('Top 3 jobs by Salary')

# Show the plot
plt.show()
```

# Top 3 countries with job vacancies


```python
#Top 3 countries with job vacancies 
top3_countries=df['company_location'].value_counts().head(3)
```

```python
top3_countries.plot(kind='barh')
plt.xlabel('count')
plt.ylabel('Countries')
plt.title('Top 3 Countries')
plt.show()
```
